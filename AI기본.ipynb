```python
import matplotlib.pyplot as plt
def boxplot_iris(feature_names, dataset):
  i=1
  plt.figure(figsize=(11,9))
  for col in feature_names:
    plt.subplot(2,2,i)
    plt.axis('on')
    plt.tick_params(axis="both", left=True, top=False, right=False, bottom=True, labelleft=False,labelbottom=False)
    dataset[col].plot(kind='box', subplots=True, sharex=False)
    plt.title(col)
    i+=1
plt.show()
boxplot_iris(iris.columns[:-1],iris)
```

```python
def historgram_iris(feature_names,dataset):
  i=1
  plt.figure(figsize=(11,9))
  for col in feature_names:
    plt.subplot(2,2,i)
    plt.axis('on')
    plt.tick_params(axis='both',left=True,top=False, right=False,bottom=False,labelleft=False,labeltop=False,labelright=False,labelbottom=False)
    dataset[col].hist()
    plt.title(col)
    i+=1
plt.show()
historgram_iris(iris.columns[:-1],iris)
```

```python
import seaborn as sns
corr=iris.corr()
cmap=sns.diverging_palette(220,10, as_cmap=True)

plt.figure(figsize=(11,9))
sns.heatmap(corr,cmap=cmap,vmax=1.0,vmin=-1.0,center=0,square=True,linewidths=.5,cbar_kws={"shrink":5})
plt.show()
```

```python
sns.pairplot(iris,hue="class")
plt.show()
```

```python
def piechart_iris(feature_names,target, dataset):
  i=1
  plt.figure(figsize=(11,9))
  for colName in [target]:
    labels=[]; sizes=[];
    df = dataset.groupby(colName).size()
    for key in df.keys():
      labels.append(key)
      sizes.append(df[key])
    plt.subplot(2,2,i)
    plt.axis('on')
    plt.tick_params(axis='both',left=False,top=False,right=False,bottom=False,labelleft=True,labeltop=True,labelright=False,labelbottom=False)
    plt.pie(sizes,labels=labels, autopct='%1.1f%%',shadow=True, startangle=140)
    plt.axis('equal')
    i+=1
  plt.show()
piechart_iris(iris.columns[:-1],iris['class'],iris)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(iris.iloc[:,:-1],iris.iloc[:,-1],test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(
    criterion='gini', splitter='best',
    max_depth=None, min_samples_split=2,
    min_samples_leaf=1, min_weight_fraction_leaf=0.0,
    max_features=None,random_state=42,
    max_leaf_nodes=None,min_impurity_decrease=0.0,
    class_weight=None
)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score, KFold
```

```python
cv = KFold(n_splits=10, shuffle=True, random_state=42)
```

```python
import numpy as np
results = cross_val_score(model,X_train,y_train,cv=cv)
fin_result = np.mean(results)
for i, _ in enumerate(results):
  print("{}번째 교차검증 정확도: {}".format(i,_))
print("\n교차검증 최종 정확도: {}".format(fin_result))
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold(n_splits=10, shuffle=True,random_state=42)
results = cross_val_score(model,X_train,y_train,cv=cv)
fin_result= np.mean(results)

for i,_ in enumerate(results):
  print("{}번째 stratified 교차검증 정확도: {}".format(i,_))
print("\nstratified 교차검증 최종 정확도: {}".format(fin_result)
)
```

```python
from sklearn.model_selection import GridSearchCV
estimator= DecisionTreeClassifier()
parameters = {'max_depth' : [4,6,8,10,12],
            'criterion' : ['gini','entropy'],
            'splitter' : ['best','random'],
            'min_weight_fraction_leaf' : [0.0,0.1,0.2,0.3],
            'random_state' : [7,23,42,78,142],
            'min_impurity_decrease' : [0.0,0.05,0.1,0.2]}
model = GridSearchCV(estimator=estimator, param_grid=parameters,cv=cv,verbose=1,n_jobs=-1,refit=True)

    
```

```python
model.fit(X_train, y_train)
```

```python
print("Best Estimator:\n",model.best_estimator_); print();
print("Best params:\n",model.best_params_);print();
print("Best Score:\n",model.best_score_)
```

```python
from sklearn.metrics import accuracy_score
pred = model.predict(X_test)
acc = accuracy_score(y_test,pred)
print('Accuracy :',acc)
```

```python
from sklearn.metrics import confusion_matrix
pred=model.predict(X_test)
confMatrix = confusion_matrix(y_test,pred)
print('Confusion Matrix: \n:',confMatrix)
```

```python
from sklearn.metrics import precision_score
precisions = precision_score(y_test,model.predict(X_test),average=None)
for target,score in zip (data['class'] ,precisions) :
  print(f"{target}의 정밀도:{score}")
```

```python
from sklearn.metrics import recall_score
recalls = recall_score(y_test,model.predict(X_test),average=None)
for target,score in zip(data['class'] ,recalls) :
  print(f"{target}의 정밀도: {score}")
```

```python
from sklearn.metrics import fbeta_score, f1_score

fbetas = fbeta_score(y_test,model.predict(X_test),beta=1,average=None)

for target,score in zip (data['class'] ,fbetas) :
  print(f"{target}의 f점수(beta=1): {score}")
print("================================")

f1s = f1_score(y_test, model.predict(X_test),average=None)

for target, score in zip  (data['class'] ,f1s) :
  print(f"{target}의 f1 점수 : {score}")

```
